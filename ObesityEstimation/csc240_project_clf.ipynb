# **Data Preprocessing**

## Handling categorical features

After exploring the dataset, I observed that converting the categorical variables into dummy variables using 'get_dummies()'. Though we don't have any strings in our dataset it is necessary to convert ('sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal') these features.

*Example: Consider the 'sex' column, it is a binary feature which has 0's and 1's as its values. Keeping it as it is would lead the algorithm to think 0 is lower value and 1 is a higher value, which should not be the case since the gender cannot be ordinal feature.*

```python
dataset = pd.get_dummies(df, columns=['Gender', 'family_history_with_overweight',
       'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS'])
```

```python
dataset.columns
```

```python
from sklearn.preprocessing import StandardScaler
standScaler = StandardScaler()
columns_to_scale = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
dataset[columns_to_scale] = standScaler.fit_transform(dataset[columns_to_scale])
```

```python
dataset
```

```python
X = dataset.drop('NObeyesdad', axis=1)
y = df['NObeyesdad']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
print('X_train size: {}, X_test size: {}'.format(X_train.shape, X_test.shape))
```

# Model Building

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

# Naive Bayes

```python
NBclf = GaussianNB()
# cross validation
cv_scores = cross_val_score(NBclf, X, y, cv=10)
print("Naive Bayes Classifier Accuracy is: {}%".format(round(cv_scores.mean(), 4)*100))
```

```python
# Test train split
NBclf = GaussianNB().fit(X_train, y_train)
NBclf.score(X_test, y_test)
```

# Decision Tree

```python
decision_scores = []
for i in range(1, 15):
  decision_classifier = DecisionTreeClassifier(max_depth=i)
  cv_scores = cross_val_score(decision_classifier, X, y, cv=10)
  decision_scores.append(round(cv_scores.mean(),3))
```

```python
# Plotting the results of decision_scores
plt.figure(figsize=(20,15))
plt.plot([i for i in range(1, 15)], decision_scores, color = 'green')
for i in range(1,15):
    plt.text(i, decision_scores[i-1], (i, decision_scores[i-1]))
plt.xticks([i for i in range(1, 15)])
plt.xlabel('Depth of Decision Tree (N)')
plt.ylabel('Scores')
plt.title('Decision Tree Classifier scores for different depth values')
```

```python
# Training the decision tree classifier model with max_depth value as 11
decision_classifier = DecisionTreeClassifier(max_depth=11)
# cross validation
cv_scores = cross_val_score(decision_classifier, X, y, cv=10)
print("Decision Tree Classifier Accuracy with max_depth = 11 is: {}%".format(round(cv_scores.mean(), 4)*100))
```

```python
# Test train split
DTclf = DecisionTreeClassifier(max_depth=11).fit(X_train, y_train)
DTclf.score(X_test, y_test)
```

# SVM

```python
# cross validation
SVMclf = svm.SVC(kernel='linear', C=1)
cv_scores = cross_val_score(SVMclf, X, y, cv=10)
print("SVM Classifier Accuracy is: {}%".format(round(cv_scores.mean(), 4)*100))
```

```python
# Test train split
SVMclf = svm.SVC(kernel='linear', C=1).fit(X_train, y_train)
SVMclf.score(X_test, y_test)
```
